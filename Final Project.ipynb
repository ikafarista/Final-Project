## Demography of Psychologists Workforce by Race from 2009 - 2019

```python
df_2019_2009 = df[(df['Year'] == 2019) | (df['Year'] == 2018)| (df['Year'] == 2017)| (df['Year'] == 2016)| (df['Year'] == 2015)| (df['Year'] == 2014)| (df['Year'] == 2013)| (df['Year'] == 2012)| (df['Year'] == 2011)| (df['Year'] == 2010)| (df['Year'] == 2009)]
df_2019_2009
```

## Psychologist Workforce Based on Race from 2009 to 2019

```python
df_2019_2009 = df_2019_2009.groupby(['Year','Race2']).sum().sum(axis=1).unstack().reset_index().set_index('Year').T
df_2019_2009
```

```python
df_2019_2009.plot.bar(figsize=(15,10))
```

## Growth of Asian American Psychologist from 2010 to 2019

```python
df_2019_2010 = df[(df['Year'] == 2019) | (df['Year'] == 2010)]
df_2019_2010
```

```python
df_2019_2010 = df_2019_2010.groupby(['Year','Race2']).sum().sum(axis=1).unstack().reset_index().set_index('Year').T
df_2019_2010
```

```python
df_2019_2010 ['pct_change'] = ((workforce_2019) - (workforce_2010))/(workforce_2019)
df_2019_2010
```
